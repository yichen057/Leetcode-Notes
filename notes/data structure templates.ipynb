```python
# 二分法模板
# 题目一般问的是: 在有序的数组中寻找一个跟target value有关的index or value. 
# 关键词: array, target, sorted, equal or close to target的词(e.g., 小于n的最大值) , O(logN), O(NlogN)即n次二分-> binary search
# 套路: 如果求两个数和的target, 固定一个值, 另一个值二分查找; 如果没有排序, 可以先排序再求解.
# 以下模板适合找第一个等于 target 的位置
# 找最左边符合条件的答案：倾向往左，最后先检查 start。
# 找最右边符合条件的答案：倾向往右，最后先检查 end。
# 二分法的时间复杂度是O(log n), n是搜索区间包含的元素个数(end - start + 1) = (n-1)-0+1=n, 也是搜索区间的长度, 即n = len(nums)
class Solution:
    def binarySearch(self, nums, target):
        if not nums:
            return -1
        
        # 如果 target 比数组最后一个数还大，应该插入到数组末尾，也就是 len(nums)
        if target > nums[-1]:
            return len(nums)
        
        # 你的模板搜索范围是这个, 也就是start 和 end 都必须是数组中真实存在的 index。
        start, end = 0, len(nums) - 1

        while start + 1 < end: # 用start+1是为了避免last position of target的情况下出现死循环: eg: nums=[1, 1], target = 1; nums[1, 2], target = 2
            # 为了统一模板, 就都采用start + 1 < end, 就保证不会出现死循环
            mid = (start + end) // 2 # python 没有overflow的问题, 直接//2 就可以了. //在python里是floor division向下取整, 包括负数也是, 而不是向0取整
            
            # 以下部分根据实际问题进行适当修改, 这里尽量不要用start = mid +/-1 or end = mid +/-1的操作
            if nums[mid] >= target:
                end = mid # 注意: 这里不要写start + 1有时会出错. 虽然 mid 已经满足条件，但左边可能还有更早的符合位置，所以不能直接返回，而要：end = mid继续想左找
                # 如果找最大值, 则需要继续向右找更大的可行值, 所以应改用start = mid
            else:
                start = mid
        # 当循环结束时, start + 1 == end, 也就是只剩下两个相邻位置：nums[start], nums[end]. 这时不要继续缩了，直接检查这两个候选答案。答案只可能是start 或end
        # 额外考虑: 因为退出条件是start + 1 < end, 因此这里循环结束的时候, start和end的关系是相邻关系, 因此需要再单独判断start和end这两个数谁是我们要的答案
        # 一般先检查start再检查end, 因为一般要求的是first index, 所以先检查更靠左的start, 如果start!=target, 再检查靠右的end
        # 最后先检查 start，因为要找靠左的答案。
        if nums[start] == target:
            return start
        if nums[end] == target:
            return end

        return -1 # 那什么时候模板里需要 return -1？ 适用于可能根本不存在符合条件答案的题。eg: LC34/704 找 target, 会有不存在的情况, 可以return -1
        # LC35找插入位置/875Koko 最小速度/1011最小船容量, 一定存在合法答案, 不需要return -1, 最后reture end即可
```

```python
def binary_search_template(nums, target):
    if not nums:
        return -1

    start, end = 0, len(nums) - 1

    while start + 1 < end:
        mid = (start + end) // 2

        # ==========================================
        # 调整点 1：根据题目定义，判断 mid 属于哪边
        #
        # 找第一个 == target: 往左找
        #     if nums[mid] >= target:
        #         end = mid
        #     else:
        #         start = mid
        #
        # 找最后一个 == target: 往右找
        #     if nums[mid] <= target:
        #         start = mid
        #     else:
        #         end = mid
        #
        # 找最大可行答案:
        #     if feasible(mid):
        #         start = mid
        #     else:
        #         end = mid
        #
        # 找 mountain peak:
        #     if nums[mid] > nums[mid + 1]:
        #         end = mid
        #     else:
        #         start = mid
        # ==========================================

    # ==========================================
    # 调整点 2：最后检查顺序
    #
    # 找最左答案：先检查 start，再检查 end
    # 找最右答案：先检查 end，再检查 start
    # 找最大值/最小值：比较 nums[start], nums[end]
    # ==========================================
```

你现在写的这个模板属于九章常用的：

while start + 1 < end：保留两个候选答案的二分模板

它最大的特点是：

循环过程中不直接把 mid 排除掉，而是让 mid 成为新的 start 或 end。
最后只剩下 start 和 end 两个候选位置，再按照题目要求检查它们。

它非常适合你当前在学的这些题：

* 找第一个 / 最后一个等于 target 的位置
* 找第一个 >= target 的位置
* 找最后一个 <= target 的位置
* 找旋转数组最小值
* 找山峰
* 在答案范围上二分，例如 Wood Cut

二、为什么循环条件是 start + 1 < end

假设：

nums = [1, 2, 2, 2, 4]
target = 2

一开始：

start = 0
end   = 4

只要中间还有没有检查完的位置，就继续二分：

while start + 1 < end:

当循环结束时：

start + 1 == end

也就是只剩下两个相邻位置：

nums[start], nums[end]

这时不要继续缩了，直接检查这两个候选答案。

例如最终可能是：

index:   0   1
nums:    1   2
       start end

因为题目要求找第一个 2，所以先判断：

if nums[start] == target:
    return start
if nums[end] == target:
    return end

三、为什么有时写 end = mid，有时写 start = mid

这里最重要的判断原则是：

mid 有没有可能是答案？

1. mid 可能是答案：保留它

例如找第一个 >= target 的位置：

if nums[mid] >= target:
    end = mid

因为 nums[mid] 已经符合条件了，它有可能就是最左边符合条件的位置，不能写成：

end = mid - 1  # 可能错误地丢掉答案

⸻

2. mid 一定不是答案：可以越过它

在另一类二分模板中，经常会看到：

start = mid + 1

例如：

while start < end:
    mid = (start + end) // 2
    if nums[mid] < target:
        start = mid + 1
    else:
        end = mid

这里之所以可以 start = mid + 1，是因为已经确定：

nums[mid] < target

所以 mid 一定不是“第一个 >= target 的位置”。

但是你现在为了统一使用九章的：

while start + 1 < end

模板，可以继续写：

start = mid

最后统一检查两个候选位置即可。

五、场景 1：普通精确查找 target

代表题目

* LeetCode 704. Binary Search

题目问：

数组中是否存在 target，存在就返回 index，不存在返回 -1。

因为不要求第一个或最后一个位置，推荐直接使用普通模板：

class Solution:
    def search(self, nums: List[int], target: int) -> int:
        start, end = 0, len(nums) - 1
        while start <= end:
            mid = (start + end) // 2
            if nums[mid] == target:
                return mid
            elif nums[mid] < target:
                start = mid + 1
            else:
                end = mid - 1
        return -1

这里调整了什么？

if nums[mid] == target:
    return mid

因为找到任意一个 target 就够了，不需要继续往左或往右找。

为什么这里用 mid + 1 / mid - 1？

因为：

nums[mid] < target

说明 mid 一定不是答案，所以可以直接排除掉。

八、场景 4：找第一个 >= target 的位置：Lower Bound

代表题目

* LeetCode 35. Search Insert Position
* 你之前九章 findUpperClosest()

题目不是要求找到等于 target 的值，而是找：

第一个大于等于 target 的位置。

例如：

nums = [1, 3, 5, 7]
target = 4

虽然 4 不在数组里，但第一个 >= 4 的数是：

5

2

更推荐的半开区间模板

这种场景我建议你逐渐掌握下面写法，因为它天然支持“插到数组最后面”的情况：

def lower_bound(nums, target):
    start, end = 0, len(nums)
    while start < end:
        mid = (start + end) // 2
        # [调整点] 找第一个 >= target
        if nums[mid] >= target:
            end = mid
        else:
            start = mid + 1
    return start

6.1 找最小值：LeetCode 153

目标：

找旋转点，也就是最小值。

关键比较对象不是 target，而是数组最后一个元素：

nums[mid] <= nums[end]

如果成立，说明 mid 位于右侧有序区间，最小值可能是 mid，也可能在左边：

end = mid

否则，说明 mid 在左侧大值区间，最小值一定在右边：

start = mid

模板

class Solution:
    def findMin(self, nums: List[int]) -> int:
        start, end = 0, len(nums) - 1
        while start + 1 < end:
            mid = (start + end) // 2
            # [调整点] 比较 nums[mid] 和 nums[end]
            if nums[mid] <= nums[end]:
                end = mid
            else:
                start = mid
        return min(nums[start], nums[end])

这里和普通查找不同：

# 普通边界题比较 target
nums[mid] >= target
# 旋转数组最小值比较右端点
nums[mid] <= nums[end]

6.2 搜索 target：LeetCode 33

这道题的关键是：

每次 mid 切开后，左半边或右半边至少有一边是有序的。
先判断哪边有序，再判断 target 是否在那边范围里。

推荐写法：

class Solution:
    def search(self, nums: List[int], target: int) -> int:
        start, end = 0, len(nums) - 1
        while start <= end:
            mid = (start + end) // 2
            if nums[mid] == target:
                return mid
            # [调整点 1] 左半边有序
            if nums[start] <= nums[mid]:
                # [调整点 2] target 是否落在左半边
                if nums[start] <= target < nums[mid]:
                    end = mid - 1
                else:
                    start = mid + 1
            # 右半边有序
            else:
                # target 是否落在右半边
                if nums[mid] < target <= nums[end]:
                    start = mid + 1
                else:
                    end = mid - 1
        return -1

这题虽然也是 binary search，但不适合机械套“第一个 target”模板，因为它需要先判断有序区间。

7.1 山脉数组唯一山顶：LeetCode 852

class Solution:
    def peakIndexInMountainArray(self, arr: List[int]) -> int:
        start, end = 0, len(arr) - 1
        while start + 1 < end:
            mid = (start + end) // 2
            # [调整点] 判断 mid 所在位置是上坡还是下坡

十二、场景 8：在答案范围上二分：Binary Search on Answer

代表题目

* LintCode 183. Wood Cut
* LeetCode 2226. Maximum Candies Allocated to K Children
* LeetCode 875. Koko Eating Bananas
* LeetCode 1011. Capacity To Ship Packages Within D Days

这类题非常重要。

它不是在数组中找一个元素，而是在一个可能答案范围里找最佳值。

例如 Wood Cut：

L = [4, 6, 7, 8]
k = 3

你要找：

每段木头长度最多是多少，才能至少切出 k 段？

可能答案范围是：

1 到 max(L)

对于一个候选长度 length，你可以判断它是否可行：

def can_cut(L, k, length):
    count = 0
    for wood in L:
        count += wood // length
    return count >= k

如果 length = 5 可行，那么更短的长度也一定可行：

1, 2, 3, 4, 5     可行
6, 7, 8            可能不可行

所以答案范围本身有单调性，可以二分。

找最大的可行答案模板

class Solution:
    def woodCut(self, L, k):
        if not L:
            return 0
        start, end = 1, max(L)
        while start + 1 < end:
            mid = (start + end) // 2
            # [调整点] mid 作为候选答案是否可行
            if self.can_cut(L, k, mid):
                # mid 可行，继续往右找更大的答案
                start = mid
            else:
                # mid 不可行，答案在左边
                end = mid
        # [调整点] 找最大的可行值，所以先检查 end
        if self.can_cut(L, k, end):
            return end
        if self.can_cut(L, k, start):
            return start
        return 0
    def can_cut(self, L, k, length):
        count = 0
        for wood in L:
            pieces = wood // length
            count += pieces
        return count >= k

这里和找最后一个 target 很像

场景	满足条件时怎么办
找最后一个等于 target 的位置	往右找
找最大的可行答案	往右找

所以：

if feasible(mid):
    start = mid
else:
    end = mid

最后先检查：

end

找到一个可能包含 target 的范围后，再二分。

如果要求找第一个 target

class Solution:
    def search(self, reader: 'ArrayReader', target: int) -> int:
        range_total = 1
        # [额外步骤] 先找到搜索范围
        while reader.get(range_total - 1) < target:
            range_total *= 2
        start, end = 0, range_total - 1
        while start + 1 < end:
            mid = (start + end) // 2
            # [和左边界模板相同]
            if reader.get(mid) >= target:
                end = mid
            else:
                start = mid
        if reader.get(start) == target:
            return start
        if reader.get(end) == target:
            return end
        return -1

这里二分部分本质仍然是：

找第一个等于 target 的位置

只是前面多了一步：

倍增寻找右边界

十五、你真正需要记住的判断规则

不要把每道题的代码都背下来。你只需要每次回答下面三个问题。

问题 1：我找的是左边答案，还是右边答案？

题目要求	倾向
第一个 target	向左
第一个 >= target	向左
最后一个 target	向右
最后一个 <= target	向右
最大可行答案	向右
最小可行答案	向左

⸻

问题 2：mid 符合条件时，能不能丢掉？

例如找第一个 target：

nums[mid] == target

mid 可能就是答案，所以：

end = mid

而不是：

end = mid - 1

例如普通查找：

nums[mid] == target

已经找到答案，可以直接：

return mid

⸻

问题 3：最后应该先检查 start 还是 end？

目标	最后检查顺序
找最左边答案	start 再 end
找最右边答案	end 再 start
找最大可行答案	end 再 start
找最小可行答案	start 再 end
找 peak / 最大值	比较两者
