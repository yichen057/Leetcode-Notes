````python
# 二分法模板
# 题目一般问的是: 在有序的数组中寻找一个跟target value有关的index or value. 
# 关键词: array, target, sorted, equal or close to target的词(e.g., 小于n的最大值) , O(logN), O(NlogN)即n次二分-> binary search
# 套路: 如果求两个数和的target, 固定一个值, 另一个值二分查找; 如果没有排序, 可以先排序再求解.
# 以下模板适合找第一个等于 target 的位置
# 找最左边符合条件的答案：倾向往左，最后先检查 start。
# 找最右边符合条件的答案：倾向往右，最后先检查 end。
# 二分法的时间复杂度是O(log n), n是搜索区间包含的元素个数(end - start + 1) = (n-1)-0+1=n, 也是搜索区间的长度, 即n = len(nums)
class Solution:
    def binarySearch(self, nums, target):
        if not nums:
            return -1
        
        # 如果 target 比数组最后一个数还大，应该插入到数组末尾，也就是 len(nums)
        if target > nums[-1]:
            return len(nums)
        
        # 你的模板搜索范围是这个, 也就是start 和 end 都必须是数组中真实存在的 index。
        start, end = 0, len(nums) - 1

        while start + 1 < end: # 用start+1是为了避免last position of target的情况下出现死循环: eg: nums=[1, 1], target = 1; nums[1, 2], target = 2
            # 为了统一模板, 就都采用start + 1 < end, 就保证不会出现死循环
            mid = (start + end) // 2 # python 没有overflow的问题, 直接//2 就可以了. //在python里是floor division向下取整, 包括负数也是, 而不是向0取整
            
            # 以下部分根据实际问题进行适当修改, 这里尽量不要用start = mid +/-1 or end = mid +/-1的操作
            if nums[mid] >= target:
                end = mid # 注意: 这里不要写start + 1有时会出错. 虽然 mid 已经满足条件，但左边可能还有更早的符合位置，所以不能直接返回，而要：end = mid继续想左找
                # 如果找最大值, 则需要继续向右找更大的可行值, 所以应改用start = mid
            else:
                start = mid
        # 当循环结束时, start + 1 == end, 也就是只剩下两个相邻位置：nums[start], nums[end]. 这时不要继续缩了，直接检查这两个候选答案。答案只可能是start 或end
        # 额外考虑: 因为退出条件是start + 1 < end, 因此这里循环结束的时候, start和end的关系是相邻关系, 因此需要再单独判断start和end这两个数谁是我们要的答案
        # 一般先检查start再检查end, 因为一般要求的是first index, 所以先检查更靠左的start, 如果start!=target, 再检查靠右的end
        # 最后先检查 start，因为要找靠左的答案。
        if nums[start] == target:
            return start
        if nums[end] == target:
            return end

        return -1 # 那什么时候模板里需要 return -1？ 适用于可能根本不存在符合条件答案的题。eg: LC34/704 找 target, 会有不存在的情况, 可以return -1
        # LC35找插入位置/875Koko 最小速度/1011最小船容量, 一定存在合法答案, 不需要return -1, 最后reture end即可
````

````python
def stringCount(self, inputStr):
    if not inputStr:
        return 0

    j, answer = 1, 0

    for i in range(len(inputStr)):
        if inputStr[i] != '0':
            continue

        j = max(j, i + 1)

        while j < len(inputStr) and inputStr[j] == '0':
            j += 1

        answer += j - i

    return answer
````

````python
# Quick Sort快排模板: 先整体有序, 再局部有序的算法(极端情况, 相同数在数组当中: [1, 1, 1, 1, 1, 2])
# Quick Sort：平均对半分，所以 O(n log n)；最坏一边倒，所以 O(n²), 但是省空间, 平均空间复杂度O(log n): 当pivot分的较均匀时, 每层总工作量都是O(n), 一共约log n层
# 如果pivot分的不够均匀, 比如pivot每次都选min or max, 而数组有序, 则递归不是对半分, 而是n, n-1, n-2..., 1, n+(n-1)+(n-2)+...+1 = (1+n)*n/2 = O(n^2)
# 即递归深度每次减少1, 最深会有n层函数调用, 因此最坏递归栈为O(n), 时间复杂度是O(n^2)
class Solution:
    def sortArray(self, nums):
        self.quick_sort(nums, 0, len(nums) - 1)
        return nums

    def quick_sort(self, nums, start, end):
        # 递归出口
        # start/end 是递归参数（描述区间）
        # left/right 是 partition指针（描述当前位置）
# 同一层递归中：
# start/end 不变
# mid 不变
# pivot 值不变
# left/right 会移动

# 下一层递归中：
# start/end 变了
# mid 重新算
# pivot 重新取
# left/right 重新初始化
        if start >= end:
            return

        left, right = start, end

        # 1. pivot get value, not index(取值，不是index), pivot不像mid没有变化, 用的是start, end计算的
        pivot = nums[(start + end) // 2]
        # 2. left <= right not left < right, 因为指针还没交叉，可以继续处理; 
        while left <= right:

            while left <= right and nums[left] < pivot:
                left += 1

            while left <= right and nums[right] > pivot: 
            # 3. nums[right] > pivot, not >=: 只跳过明确比 pivot 大的. 不要写成nums[right] >= pivot, 因为等于 pivot 的值也需要停下来处理
                right -= 1

            if left <= right:
                nums[left], nums[right] = nums[right], nums[left]
                left += 1
                right -= 1
        # 当left > right时 partition结束. left/right 真正有意义的位置，是 partition 结束后的 left 和 right。

        # 递归左右两边
        # 这里的 left 和 right 是移动后的值，不是交换前的位置。
        self.quick_sort(nums, start, right)
        self.quick_sort(nums, left, end)
````

````python
# MergeSort 归并排序模板: 先局部有序, 再整体有序的算法. 先拆到最小，再一层层合并有序数组
# Merge Sort：永远对半分，层数一定是log n, 每一层merge共处理所有元素一次, 即每层O(n), 所以时间复杂度稳定 O(n log n), 但需要额外O(n)的空间: temp: O(n) + 递归栈:O(log n) = O(n)
# 和快速排序比的劣势: 需要花费O(n)的额外空间, 为了合并两个数组(无法原地合并)
# merge_sort(nums, start, mid)
# merge_sort(nums, mid + 1, end)
# merge(nums, start, mid, end)

# Time complexity: O(nlogn)
# We divide the nums array into two halves till there is only one element in the array, which will lead to O(logn) steps.问题规模每次都除以2, 除几次会变成1
# n→n/2→n/4→...→1 (k steps)
# n/2 ^(k−1) =1 ⟹ 2^k = n ⟹ k≈logn, 所以递归深度是O(log n)
# And after each division, we merge those respective halves which will take O(n) time each.
# Thus, overall it takes O(nlogn) time.

# Space complexity: O(n)
# The recursive stack will take O(logn) space and we used an additional array temporaryArray of size n.
# Thus, overall we use O(logn+n)=O(n) space.

# sortArray()
#     ↓
# merge_sort()
#     ↓
# merge_sort()
#     ↓
# merge()
# 所有函数都在操作同一个：nums

# 参考LC 912
class Solution:
    def sortArray(self, nums):
        if not nums:
            return nums

        temp = [0] * len(nums)
        self.merge_sort(nums, 0, len(nums) - 1, temp)
        return nums

    # merge_sort用来拆分, 排序左右半边
    def merge_sort(self, nums, start, end, temp): # in-place modification
        # 递归出口
        if start >= end:
            return

        mid = (start + end) // 2

        # 先排序左半边
        self.merge_sort(nums, start, mid, temp)

        # 再排序右半边
        self.merge_sort(nums, mid + 1, end, temp)

        # 合并两个有序区间
        self.merge(nums, start, mid, end, temp)
    
    # merge: 用来合并两个有序数组. 进入 merge() 时：整个数组不一定有序, 但是左半边一定有序, 右半边一定有序
    def merge(self, nums, start, mid, end, temp): # in-place modification
        left = start
        right = mid + 1
        index = start # 指的是temp的index, 初始化为start

        # 合并两个 sorted subarrays
        while left <= mid and right <= end:
            if nums[left] <= nums[right]: # 如果想让归并排序保持 stable sort 稳定性，这里的<可以写成<=. 刷 LeetCode 排整数，两种都能过。稳定排序场景下更推荐 <=。
                temp[index] = nums[left]
                left += 1
            else:
                temp[index] = nums[right]
                right += 1 # 注意: 此处r也是向右移动+1, 而不是从末尾往左走-1
            index += 1

        # 左边还有剩余, 左边可能剩 O(n) 个, 所以用while而不是if
        while left <= mid:
            temp[index] = nums[left]
            left += 1
            index += 1

        # 右边还有剩余, 右边也可能剩 O(n) 个, 所以用while而不是if
        while right <= end:
            temp[index] = nums[right]
            right += 1
            index += 1

        # copy temp back to nums, 这里实际上已经修改了nums, 所以没必要再return nums
        for i in range(start, end + 1): # 注意这里的区间[start, end + 1)
            nums[i] = temp[i]
````

````python
# quick select template
# partition() 只递归一边
# quick select: 平均时间复杂度O(n), 最差时间复杂度O(n^2), 空间复杂度O(logn), 只递归包含答案的range, 不断partition缩小范围: 三个范围: >pivot, <pivot, ==pivot
# 而quickl sort两边都找, >=pivot, <=pivot
# Quick Select 比 Quick Sort 快. 因为它只递归一边, 而不是两边都递归
# quick select代码框架
def findKthLargest():

    quickSelect()

        partition

        target = start + k - 1 # 第k大元素在当前区间中的目标下标

        if target <= right:
            go left

        elif target >= left: # 此时 k =  k - (left - start), 表示把全局第k大转换成右边区间里的第几大
            go right

        else:
            return nums[right+1]

# 具体模板, 参考LC 215题
class Solution:
    def findKthLargest(self, nums: List[int], k: int) -> int:
        if not nums:
            return -1
        return self.quickSelect(nums, 0, len(nums)-1, k)

    def quickSelect(self, nums: List[int], start: int, end: int, k: int) -> int:
        if start == end:
            return nums[start] # 注意: 这里返回的不是index: start. 而是返回元素: nums[start]
        
        left = start
        right = end
        pivot = nums[(left+right)//2]

        while left <= right:
            while left <= right and nums[left] > pivot:# 此处从大小排序, 注意这里的左边区间是大于pivot而不是小于pivot. 这里是降序数组
                left += 1
            while left <= right and nums[right] < pivot:
                right -= 1
            
            if left <= right: # 注: 这个 if left <= right: 应该在外层 while left <= right: 里面。否则如果 nums[left] 和 nums[right] 都不能移动，外层 while 会一直卡住，导致 TLE(time limit exceeded)。
                nums[left], nums[right] = nums[right], nums[left]
                left += 1
                right -= 1
        # 此时partition结束, right > left, 分区后有三个区域: 下面的target指的是第k大元素的index
        # [start...right]            > pivot    当target <= right, 答案在right左边
        # [right + 1...left - 1]    == pivot    当right < target < left, 答案在==pivot中间
        # [left...end]               < pivot    当target >= left, 答案在left右边

        # 降序数组中的第k大的index: target = start + k - 1
        target = start + k - 1
        # 1) > pivot区间, 左边
        if target <= right:# 注意: 这里是小于等于号, 并非小于号
            return self.quickSelect(nums, start, right, k)
        # 2) < pivot区间, 右边
        if target >= left:
            return self.quickSelect(nums, left, end, k - (left-start)) 
            # 这里的k-(left-start)是将全局排名转为局域排名: k是在全局数组里的第k大, 转为右边局域区间数组的第[k-(left-start)]大
        # 3) == pivot区间, 中间
        return nums[right + 1] # right + 1是 == pivot区域内的第一个元素的index

````

Quick Sort：
先 partition，把小的放左边，大的放右边，再递归左右区间。

````python
# Merge Interval 模板 (LC56)
class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        intervals.sort(key = lambda x: x[0]) # 时间复杂度: O(n log n)
        # 先排序的原因是: 按每个区间的开始值从小到大排。排好之后,有个很爽的保证:能重叠的区间一定是挨着的。所以我们只需要拿"新区间"和"结果里最后一个区间"比较就够了,不用回头看前面的。
        output = [intervals[0]] # 这是一个装着区间的列表. 为什么要先把第一个区间塞进去?因为后面的逻辑是"拿新区间去和结果里最后一个比",总得先有一个东西可比吧,所以把第一个当"种子"先放进去。
        for start, end in intervals[1:]: # 时间复杂度: O(n)
            lastEnd = output[-1][1] # 结果列表里的最后一个区间, [1] = 这个区间的结束值/右端点(索引 0 是开始,1 是结束)
            # 排序保证了"当前区间的 start 比已处理的所有区间都大(或相等), 所以排完序之后,只有"最后一个"才可能和当前区间重叠
            # 当判断出重叠时(start <= lastEnd),要把最后一个区间的右端点往右伸长。
            # 判断重叠的完整条件: newStart <= lastEnd and lastStart <= newEnd, 本题因为排序后lastStart<=newStart<= newEnd, 所以只需判断newStart <= lastEnd
            if start <= lastEnd:
                output[-1][1] = max(end, lastEnd) # 为什么用 max 而不是直接 = end? 因为新区间有可能整个被"包"在里面
            else:
                output.append([start, end])
        return output
````

````python
# Insert interval (LC57)
# method 2: 三段式扫描数组(面试和刷题更推荐 Method 2。因为intervals 按 start 升序排列，且彼此不重叠)
# 时间复杂度：O(n): 因为Scan all intervals once: O(n)
# 空间复杂度：O(n)。因为Result array: O(n), 如果不算输出数组，额外空间可以认为是 O(1)。
# 左边区间->重叠区间->右边区间, 三个连续区域
# 第一阶段(收集所有在 newInterval 左边、且不重叠的区间): i < n and intervals[i][1] < newInterval[0]:只要还在左边区域, 就一直收集
# 第一阶段结束时, i自动停在第二阶段起点, 第一阶段结束后已经保证：curEnd >= newStart, 即newStart <= curEnd
# 第二阶段(merge所有和 newInterval 重叠的区间): while i < n and intervals[i][0] <= newInterval[1]:
# 第三阶段(收集所有在 newInterval 右边、且不重叠的区间): while i < n:
# eg: [1,2] | [3,5] [6,7] [8,10] | [12,16]
           #与newInterval[4, 8]重叠

# 判断重叠的完整条件: newStart <= lastEnd and lastStart <= newEnd
# 此处用while是因为需要一直往前走, 直到条件失效; 而for循环是用来把所有元素都看一遍
# 另外注意本题的intervals是按start升序排列
class Solution:
    def insert(self, intervals: List[List[int]], newInterval: List[int]) -> List[List[int]]:
        res = []
        i = 0
        n = len(intervals)
        # 1. 收集所有在 newInterval 左边、且不重叠的区间
        while i < n and intervals[i][1] < newInterval[0]:
            res.append(intervals[i])
            i += 1
        # 2. 合并所有和 newInterval 重叠的区间
        # 此时i自动停在第二阶段起点, 第一阶段结束后已经保证：lastEnd >= newStart, 即newStart <= lastEnd, 所以判断重叠区间只需判断另外一部分: lastStart <= newEnd:
        while i < n and intervals[i][0] <= newInterval[1]:
            newInterval[0] = min(newInterval[0], intervals[i][0])
            newInterval[1] = max(newInterval[1], intervals[i][1])
            i += 1
        res.append(newInterval)

        # 3. 收集所有在 newInterval 右边、且不重叠的区间
        while i< n:
            res.append(intervals[i])
            i += 1
        
        return res
    
````

以下是heap的基础知识
Python 里的 heapq 默认是 min heap（最小堆）, heap[0]永远是当前 heap 里的最小值, TC: O(1)
heapq.heapify(): 将list原地转为heap, TC: O(n)
heappush(): heap 添加元素并保持heap结构, TC: O(log n)
heappop(): heap弹出并返回最小值, smallest = heapq.heappop(heap), TC: O(log n)

````python
import heapq

nums = [5, 2, 8, 1, 3]
heapq.heapify(nums)

print(nums) # nums不排序, 但是保证nums[0]最小
````

````python
# Min-Heap method with k largest integers (size k)
# 本题中的求最大并不是找max heap, 我们真正需要的是Top k largest 中最小的那个, 即第k大, 所以heap[0]就能拿到最小值
# init: TC: O(n log n)
# TC: O(n logk), Because k≤n, this is an improvement on the previous approach.
# SC: O(k), because the heap uses O(k) space
import heapq
class Solution:
    def findKthLargest(self, nums: List[int], k: int) -> int:
        heap = [] # SC: O(k) because the heap uses O(k) space
        for num in nums: # iterate n times, performing up to (log k) work at each iteration -> TC: O(n logk )
            heapq.heappush(heap, num)
            if len(heap) > k:
                heapq.heappop(heap)
        return heap[0]
````
